````python
# Your code module 6, video 4: 
# #### Part A ######
def aggregate_data(dfx):
    ######## PART B #######
    # part I: we don't run this if the threads column is not in the dataframe. 
    if "threads" in dfx.columns:
        # creating two dataframes one for sum and one for count. 
        # as shown in the lect, returns the sum of the threads as values
        threads_sum = dfx.pivot_table (index = "image", values = "threads", aggfunc = "sum")
        # count is from the pandas library, counts all values excluding the NaN / NA values. 
        process_qty = dfx.pivot_table (index = "image", values = "threads", 
                                       aggfunc = "count")

        # resetting the index: from online search.
        threads_sum = threads_sum.reset_index()
        process_qty = process_qty.reset_index()

#     # merging the two dataframes such that it only looks at the column image 
#     as key column and not other columns. Without ON there maybe erros in the result. 
        df_merger = pd.merge(process_qty, threads_sum, on = "image")

    # Rename the columns based on the instrucitons:
        df_merger.rename({"image" : "image_name", "threads_x" : "process_qty",
                  "threads_y" : "threads_sum" }, axis = 1, inplace=True)
        
#     Part II 
    elif "total_byte_per_second" in dfx.columns:

        total_byte_per_second_sum = \
        dfx.pivot_table (index = "image", 
                             values = "total_byte_per_second", aggfunc = "sum")
        process_qty = dfx.pivot_table (index = \
                                           "image", values = "total_byte_per_second", 
                                           aggfunc = "count")

        # resetting the index: from online search.
        total_byte_per_second_sum = total_byte_per_second_sum.reset_index()
        process_qty = process_qty.reset_index()

        # merging the two dataframes such that it only looks at the column image 
        # as key column and not other columns.
        df_merger = pd.merge(process_qty, total_byte_per_second_sum, on = "image")

        # Renaiming the columns based on the instrucitons:
        df_merger.rename({"image" : "image_name", "total_byte_per_second_x" : 
                          "process_qty", "total_byte_per_second_y" : 
                          "total_byte_sum" }, axis = 1, inplace = True)
        
#     Part III    
    elif "working_set_kb" in dfx.columns:
        working_set_kb_sum = \
        dfx.pivot_table (index = "image", values = "working_set_kb", aggfunc = "sum")
       # count is from the pandas library, it returns all 
       # values in a row except the NaN values, etc.
        process_qty = \
        dfx.pivot_table (index = "image", values = "working_set_kb", aggfunc = "count")

        # resetting the index: from online search.
        working_set_kb_sum = working_set_kb_sum.reset_index()
        process_qty = process_qty.reset_index()

        # merging the two dataframes such that it 
        # only looks at the column image as key column and not other columns.
        df_merger = pd.merge(process_qty, working_set_kb_sum, on = "image")

    # Renaiming the columns based on the instrucitons:
        df_merger.rename({"image" : "image_name", "working_set_kb_x" :
                          "process_qty", "working_set_kb_y" : "working_set_sum" }, 
                         axis = 1, inplace = True)

    return df_merger


# Setting variables
df_cpu_sum = aggregate_data(df_cpu)
df_disk_sum = aggregate_data(df_disk)
df_memory_sum = aggregate_data(df_memory)

# print(df_cpu_sum.head(10))
print(f"df_cpu_sum: HEAD \n\n {df_cpu_sum.head(10)}\n\n")
print(f"df_cpu_sum: TAIL \n\n {df_cpu_sum.tail(10)}\n\n")


print(f"df_disk_sum: HEAD \n\n {df_disk_sum.head(10)}\n\n")
print(f"df_disk_sum: TAIL \n\n {df_disk_sum.tail(10)}\n\n")


print(f"df_memory_sum: HEAD \n\n {df_memory_sum.head(10)}\n\n")
print(f"df_memory_sum: TAIL \n\n {df_memory_sum.tail(10)}\n\n")

print(f"the shape of df_cpu_sum is : {df_cpu_sum.shape}")
print(f"the shape of df_disk_sum is : {df_disk_sum.shape}")
print(f"the shape of df_memory_sum is : {df_memory_sum.shape}")

````
